# Generate figure for poster/report with the four input types.

```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
# Choose a file
y, sr = librosa.load('archive/Train_submission/Train_Submission/dramatic-piano-10950.wav')

fig, ax = plt.subplots(2,2)

# Display spectrograms
S = librosa.feature.melspectrogram(y=y, sr=sr)
librosa.display.specshow(librosa.power_to_db(S, ref=np.max), ax=ax[0,0], y_axis='mel', x_axis='time')

chroma = librosa.feature.chroma_cqt(y=y, sr=sr)
librosa.display.specshow(chroma, y_axis='chroma', x_axis='time', ax=ax[0,1])

S = librosa.feature.melspectrogram(y=y, sr=sr, n_fft=512)
librosa.display.specshow(librosa.power_to_db(S, ref=np.max), ax=ax[1,0], y_axis='mel', x_axis='time')

librosa.display.waveshow(y,sr=sr, axis='time', color='blue', ax=ax[1,1])

plt.tight_layout()
```
